```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
image = plt.imread('download.jpeg')
```

```python
plt.imshow(image)
```

```python
plt.imshow(image[: ,100:350])
image.shape
```

```python
order = np.reshape(image , [image.shape[0]*image.shape[1] ,-1])
order.shape
```

```python
order
```

```python
kmeans = KMeans(n_clusters = 16 ,random_state = 0)
kmeans = kmeans.fit(order)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
labels = centroids[labels]
labels
```

```python
len(order) == len(labels)
```

```python
order2 = order.copy()
```

```python
for i in range(len(order)):
    order2[i] = labels[i] 
    
```

```python
comp_image = np.reshape(order2 ,image.shape)
```

```python
plt.imshow(comp_image)
```
