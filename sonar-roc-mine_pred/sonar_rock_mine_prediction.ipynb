```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score    ## for measuring accuracy of model
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Desktop\New folder\Copy of sonarr data.csv",header = None )
```

```python
X = df.iloc[: , :60]
Y = df.iloc[: , 60]
```

```python
X_train , X_test , Y_train, Y_test = train_test_split(X , Y , test_size = 0.1 , stratify = Y , random_state = 1)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

```python
model = LogisticRegression()
model.fit(X_train , Y_train)
```

```python
X_train_predict = model.predict(X_train)
train_acc_score = accuracy_score(X_train_predict , Y_train)
```

```python
X_test_predict = model.predict(X_test)
test_acc_score = accuracy_score(X_test_predict , Y_test)
```

```python
print(f" training data accuracy score: {train_acc_score}  \n testing data accuracy score: {test_acc_score}")
```
